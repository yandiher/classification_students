```python
x = df.drop(['name', 'label'], axis=1).values
y = df.label.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, stratify=y)
```

```python
import statsmodels.api as sm
```

```python
param = ['constant', 'umur', 'kemampuan bahasa inggris','tingkat sekolah']
```

```python
X_train = sm.add_constant(x_train)
train_model = sm.OLS(exog=X_train, endog=y_train).fit()
```

```python
print(train_model.summary(xname=param))
```

```python
X_test = sm.add_constant(x_test)
test_model = sm.OLS(exog=X_test, endog=y_test).fit()
```

```python
print(test_model.summary(xname=param))
```

```python
X = sm.add_constant(x)
model = sm.OLS(exog=X, endog=y).fit()
```

```python
print(model.summary(xname=param))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
print('the score :', r2_score(y_true=y_test, y_pred=pred).round(2))
```

```python
def kelas(kelas):
    if kelas == 8:
        return 'Diligent'
    elif kelas == 7:
        return 'Empathy'
    elif kelas == 6:
        return 'Integrity'
    elif kelas == 5:
        return 'Responsible'
    elif kelas == 4:
        return 'Creativity'
    elif kelas == 3:
        return 'Tolerance'
    elif kelas == 2:
        return 'Honesty'
    elif kelas == 1:
        return 'Enthusiastic'
    elif kelas == 0:
        return 'ngaco ge. coba cek lagi.'
    else:
        return 'cek lagi geura. keknya ngaco.'
```
