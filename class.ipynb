# data algorithm

```python
x = df[['your age', 'total']]
y = df['your class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, train_size=0.8, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
accuracy_score(y_true=y_test, y_pred=pred).round(2)
```

```python
print(classification_report(y_true=y_test, y_pred=pred))
```

```python
ConfusionMatrixDisplay.from_estimator(X=x, y=y, estimator=clf, labels=order, xticks_rotation=45);
```
